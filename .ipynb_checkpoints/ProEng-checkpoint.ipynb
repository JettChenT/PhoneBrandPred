```python
brands = [
    "Apple",
    "Google Pixel",
    "XiaoMi",
    "Samsung",
    "Oppo",
    "Vivo",
    "8848"
]
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a.lower(), b.lower()).ratio()
```

```python
def getsim(brand):
    maxi = -1
    maxs = -1
    for i in range(len(brands)):
        s = similar(brand,brands[i])
        if s>maxs:
            maxs = s
            maxi = i
    if maxs<0.3:
        return -1
    return maxi
```

```python
def psn(brand):
    n = getsim(brand)
    if n==-1:
        return "other"
    return brands[n]
```

```python
psn("8848 vibranium")
```

```python
data["brand"] = [psn(b) for b in data["brand"]]
```

```python
data["prefBrand"] = [psn(b) for b in data["prefBrand"]]
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
acnt = [0]*(len(brands)+1)
for bnd in data['brand']:
    acnt[getsim(bnd)]+=1
print(acnt)
```

```python
def clean_data(ldat,lbel):
    lbel.append("Others")
    rdat,rbel = [],[]
    for i in range(len(ldat)):
        if ldat[i]!=0:
            rdat.append(ldat[i])
            rbel.append(lbel[i])
    return rdat,rbel
```

```python
ccnt,ccrand = clean_data(acnt,brands)
fig, ax = plt.subplots(figsize=(10,10))
ax.pie(ccnt, labels = ccrand,autopct='%.0f%%')
ax.set_title('Phone usage Brands')
ax.legend(loc='best')
plt.show()
```

```python
reasons = [
    "Memory",
    "Pricing",
    "Storage",
    "Operating system",
    "Brand",
    "Publicity",
    "Screen resolution",
    "Screen refresh rate",
    "Screen size",
    "Battery",
    "Other"
]
```

```python
ResCode = [
    "Mem",
    "Pric",
    "Storage",
    "OS",
    "Brand",
    "Pub",
    "ScrRes",
    "RefRate",
    "ScrSize",
    "Battery",
    "Other"
]
```

```python
def get_rs(rstr):
    res = [0]*len(reasons)
    rsns = rstr[:-1].split(";")
    for i in range(len(reasons)-1):
        if reasons[i] in rsns:
            res[i] = 1
    for r in rsns:
        if r not in reasons:
            res[-1]=1
    return res
```

```python
get_rs("Battery;Memory;Storage;Operating system")
```

```python
fctrs = np.zeros((len(data),len(reasons)),dtype=np.int32)
for i in range(len(data['factor'])):
    rsns = get_rs(data["factor"][i])
    fctrs[i] = rsns
```

```python
for i in range(len(reasons)):
    data[f"Res{ResCode[i]}"] = fctrs[:,i]
```
